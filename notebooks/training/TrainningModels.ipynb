```python
import pandas as pd
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
df = pd.read_csv('../datasets/ground.csv')
# # Táº¡o training data
training_df =pd.read_csv('../datasets/ground.csv')

# # Separate X, y
feature_cols = [col for col in training_df.columns 
               if col not in ['label', 'query', 'venue_id', 'query_id']]
X_train = training_df[feature_cols]
y_train = training_df['label']
X_test = X_train.iloc[:50]
y_test = y_train.iloc[:50]
# # Train model
dummy = DummyClassifier(strategy='most_frequent')
dummy.fit(X_train, y_train)
y_pred = dummy.predict(X_test)
print(f"Baseline Accuracy: {accuracy_score(y_test, y_pred)}")
```

```python
import joblib

joblib.dump(dummy, "../checkpoints/dummy.pkl")
```

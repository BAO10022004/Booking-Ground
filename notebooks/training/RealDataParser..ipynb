```python
import json
from pathlib import Path

def load_json(file_path, encoding='utf-8', verbose=True, extract_key=None):
    try:
        file_path = Path(file_path)
        
        if verbose:
            print(f"üìÇ Loading: {file_path}")
        
        if not file_path.exists():
            return None
        
        with open(file_path, 'r', encoding=encoding) as f:
            data = json.load(f)
        if extract_key:
            if isinstance(data, dict) and extract_key in data:
                if verbose:
                    print(f"üîë Extracting key: '{extract_key}'")
                data = data[extract_key]
            else:
                print(f"‚ö†Ô∏è Key '{extract_key}' not found in JSON")
        if verbose:
            data_type = type(data).__name__
            
            if isinstance(data, list):
                print(f"‚úÖ Loaded {len(data)} items (list)")
            elif isinstance(data, dict):
                print(f"‚úÖ Loaded dictionary with keys: {list(data.keys())}")
                # N·∫øu dict c√≥ nested list, show th√™m th√¥ng tin
                for key, value in data.items():
                    if isinstance(value, list):
                        print(f"   - {key}: {len(value)} items")
            else:
                print(f"‚úÖ Loaded {data_type}")
        
        return data
        
    except json.JSONDecodeError as e:
        return None
    except Exception as e:
        return None
```

```python
data_venue = load_json('../data/raw/venue.json')
df_venue = pd.DataFrame(data_venue)
df_venue = df_venue['venues'].tolist()
df_venue = pd.json_normalize(df_venue)
print(df_venue)
```

```python
# Drop column 

df_venue = pd.DataFrame(df_venue)
cols_to_drop = ['name', 'website', 'ownerId', 'image', 'phoneNumber2', 'address']
df_venue.drop(columns=cols_to_drop, inplace=True)
print(df_venue)
```

```python
df_ground = load_json('../data/raw/ground.json')
df_ground = pd.DataFrame(df_ground)
df_ground = df_ground['grounds'].tolist()
df_ground = pd.json_normalize(df_ground)
print(df_ground)
```

```python
df_ground = df_ground.drop(columns=['name'])
print(df_ground)
```

```python
df_category = load_json('../data/raw/categories.json')
df_category = pd.DataFrame(df_category)
df_category = df_category['categories'].tolist()
df_category = pd.json_normalize(df_category)
print(df_category)
```

```python
df_ground = pd.merge(
    df_ground,
    df_category,
    on='categoryId',  
    how='left'          

)
print(df_ground)
df_ground = df_ground.drop(columns=['categoryId'])
```

```python
df = pd.merge(
    df_venue,
    df_ground,
    on='venueId',
    how='left'
)
print(df)
```

```python
df = df.rename(columns={
    'name': 'category',      # T√™n c·ªßa s√¢n l·ªõn/ƒë·ªãa ƒëi·ªÉm
})

df[['start_time', 'end_time']] = (
    df['operatingTime']
    .str.split('-', expand=True)
    .apply(lambda col: col.str.strip())
)
df = df.drop(columns=['phoneNumber1'])
df = df.drop(columns=['operatingTime'])
print(df)
```

```python
df = df.drop(columns=['deposit','groundId' ])
print(df)
```
